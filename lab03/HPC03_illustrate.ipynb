```python
import matplotlib.pyplot as plt
import numpy as np
# Define the fraction of serial work
f_serial = 0.45
# Define the number of processors
n_processors = np.array([1, 2, 3, 4, 5, 6, 7, 8])
# Define the speedup values
speedup_actual = np.array([1, 
                           1.569444,
                           1.603008,
                           1.913117,
                           1.608680,
                           1.498681,
                           1.264425,
                           1.572610])
# Calculate the theoretical speedup based on Amdahl's Law
speedup_theoretical = 1 / (f_serial + ((1 - f_serial) / n_processors))
# Plot the actual and theoretical speedup graphs
plt.plot(n_processors, speedup_actual, 'ro-', label='Реальне прискорення')
plt.plot(n_processors, speedup_theoretical, 'b*-', label='Теоретичне прискорення')
# Set the plot title and axis labels
plt.title("Реальне та теоретичне прискорення, базовані на законі Амдала")
plt.xlabel("Кількість процесорів")
plt.ylabel("Прискорення")
# Add a legend to the plot
plt.legend()
# Show the plot
plt.show()
```
